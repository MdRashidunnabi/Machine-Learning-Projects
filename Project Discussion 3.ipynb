# Yeo Jonson Method

```python
#Prepare the dataset for training the model
x=new_diabetes.drop('Outcome',axis=1)
y=new_diabetes['Outcome']
```

Since the outcome has only two variables we will use binary classification model.

# Yeo-Jonson Method

```python
from sklearn.preprocessing import power_transform     #box-cox is only for positive but yeo-johnson is for both negative and positive
x=power_transform(x,method='yeo-johnson')
```

Since the outcome has only two variable we will use binary classification model

1. identify best random_state using Logistic Regression

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_t=sc.fit_transform(x)
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
from sklearn.metrics import accuracy_score
```

```python
max_scr=0
for i in range(0,100):
    x_train,x_test,y_train,y_test=train_test_split(x_t,y,test_size=.42,random_state=i)
    lr.fit(x_train,y_train)
    pred_train=lr.predict(x_train)
    pred_test=lr.predict(x_test)
    if round(accuracy_score(y_train,pred_train)*100,2)==round(accuracy_score(y_test,pred_test)*100,2):
        print("At random state",i,"The model performs very well")
        print("At random state :-",i)
        print("Training accuracy_score is :-",accuracy_score(y_train,pred_train)*100)
        print("Testing accuracy_score is :-",accuracy_score(y_test,pred_test)*100)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_t,y,test_size=.42,random_state=90)
```

# Cross Valodation score for logistic Regression

```python
pred_lr=lr.predict(x_test)
from sklearn.model_selection import cross_val_score
lss=accuracy_score(y_test,pred_lr)
for j in range(2,20):
    lsscore=cross_val_score(lr,x,y,cv=j)
    lsc=lsscore.mean()
    print("At cv:-",j)
    print("Cross validation score is :- ", lsc*100)
    print("accuracy_score is :-",lss*100)
    print("\n")
```

Since the cross validation score and accuracy score are almost same .We are heading with a good approach

```python
lsscore_selected=cross_val_score(lr,x,y,cv=16).mean()
print("the cv score is : ",lsscore_selected,"\nThe accuracy score is:",lss)
```

# AUC ROC Curve

```python
from sklearn.metrics import roc_curve,auc
fpr,tpr,thresholds=roc_curve(pred_test,y_test)
roc_auc=auc(fpr,tpr)

plt.figure()
plt.plot(fpr,tpr,color='darkorange',lw=10,label='Roc curve (area=%0.2f)'%roc_auc)
plt.plot([0,1],[0,1],color='navy',lw=10,linestyle='--')
plt.xlim([0.0,1.5])
plt.ylim([0.0,1.5])
plt.xlabel('False positive rate')
plt.ylabel('True Positive rate')
plt.title('Receiver operating  characteristic')
plt.legend(loc="lower right")
plt.show()
```

# Hyperparameter Tuning

```python
#Creating parameter list to pass the GridSearchCV
parameters = {"max_iter":[500,700,900,1100,1300,1400,1500],     #maximum iteration.It means how many times I will pass the data for iteration.
              "penalty" : ['l1','l2','elasticnet','none']}           #The best parameter will be selected automatically. We can use this code everywhere
```

```python
from sklearn.model_selection import GridSearchCV
GCV=GridSearchCV(LogisticRegression(),parameters,cv=16,scoring='accuracy')      #Initializing GridSearchCV cv=cross_validation ,r2 is for improving the accuracy
GCV.fit(x_train,y_train)                                           #cv=5 means 5 cross folds will be created and there will be training and testing data . Any fold will not be empty.
GCV.best_estimator_  #finding the best estimator
GCV_pred=GCV.best_estimator_.predict(x_test) #Predicting the value using best estimator found by GridSearchCV
print("Final accuracy : ",accuracy_score(y_test,GCV_pred)*100) #Final accuracy
```

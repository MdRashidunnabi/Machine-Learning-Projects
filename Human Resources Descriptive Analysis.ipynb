# Education

```python
attrition_EducationField = pd.DataFrame(columns=["Field", "% of Leavers"])
i=0
for field in list(attrition['EducationField'].unique()):
    ratio = attrition[(attrition['EducationField']==field)&(attrition['Attrition']=="Yes")].shape[0] / attrition[attrition['EducationField']==field].shape[0]
    attrition_EducationField.loc[i] = (field, ratio*100)
    i += 1    
    #print("In {}, the ratio of leavers is {:.2f}%".format(field, ratio*100))    
attrition_EF = attrition_EducationField.groupby(by="Field").sum()
attrition_EF.plot(kind='bar',title='Leavers by Education Field (%)',figsize=(12,10))
```
